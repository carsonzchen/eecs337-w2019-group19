```python
#stopWords = set(stopwords.words('english'))
stopWords = ["http", "rt", "goldenglobes", "golden", "globes", "RT", "Golden", "Globes", "GoldenGlobes"]
keywords = ['winner', 'award', 'nominees', 'congratulations', 'announce']
```

```python
# tokenize tweets (reference to GG_Presenters_CarsonChen.ipynb. modified a bit)
start = len(data)-200
end = len(data)
word_list = []
for q in range(start, end):
    words = word_tokenize(data[q]['text'])
    words_meaningful = [token for token in words if token not in stopWords and token.isalpha()]
    if len(list(set(words_meaningful) & set(keywords)))>0:
        word_list.append(words_meaningful)
        

for q in word_list:
    print(q, '\n')
```
