```python
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk import bigrams, trigrams
import json
from pprint import pprint
import os
```

### Bigrams & Trigrams
#### Traditional Loops

```python
# load json data
dir_path = os.path.dirname(os.path.realpath(""))
with open(dir_path + '/eecs337-w2019-group19/data/gg2013.json') as f:
    data = json.load(f)

counter = 0
common_bigrams = {}
common_trigrams = {}
stopWords = set(stopwords.words('english'))
twitterwords = {"http", "rt", "goldenglobes", "globes"}
stopWords = stopWords.union(twitterwords)
```

for tweet in data:
    #pprint(tweet['text'])
    counter+=1
    words = word_tokenize(tweet['text'].lower())
    wordstop = [token for token in words if token not in stopWords and token.isalpha()]
    bigrm_in_tweet = list(bigrams(wordstop))
    trigrm_in_tweet = list(trigrams(wordstop))
    
    for bg in bigrm_in_tweet:
        if bg in common_bigrams.keys():
            common_bigrams[bg] += 1
        else:
            common_bigrams[bg] = 1
    
    for tg in trigrm_in_tweet:
        if tg in common_trigrams.keys():
            common_trigrams[tg] += 1
        else:
            common_trigrams[tg] = 1
    
    if counter > 100:
        break

### Combined tweets

```python
tweettext = []
counter = 0

for tweet in data:
    #pprint(tweet['text'])
    counter += 1
    words = word_tokenize(tweet['text'].lower())
    wordstop = [token for token in words if token not in stopWords and token.isalpha()]
    tweettext.append(wordstop)
    
    if counter >= 155:
        break
```

```python
for sentence in tweettext:
    bigrm_in_tweet = list(bigrams(sentence))
    for bg in bigrm_in_tweet:
        if bg in common_bigrams.keys():
            common_bigrams[bg] += 1
        else:
            common_bigrams[bg] = 1
```

```python
for sentence in tweettext:
    trigrm_in_tweet = list(trigrams(sentence))
    for tg in trigrm_in_tweet:
        if tg in common_trigrams.keys():
            common_trigrams[tg] += 1
        else:
            common_trigrams[tg] = 1
```

### Show results

```python
d_view = [ (v,k) for k,v in common_bigrams.items() ]
d_view.sort(reverse=True) # natively sort tuples by first element
counter = 0
for v,k in d_view:
    pprint("%s: %d" % (k,v))
    counter+=1
    
    if counter > 500:
        break
```

```python
d_view = [ (v,k) for k,v in common_trigrams.items() ]
d_view.sort(reverse=True) # natively sort tuples by first element
counter = 0
for v,k in d_view:
    pprint("%s: %d" % (k,v))
    counter+=1
    
    if counter > 100:
        break
```
